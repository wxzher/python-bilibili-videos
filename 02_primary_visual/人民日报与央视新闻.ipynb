```python
# 导入需要的工具包
import matplotlib.pyplot as plt
from matplotlib import font_manager        #导入字体管理模块
import pandas as pd

```

```python
# 1.读取数据
# 使用pandas读取表格数据
df_renmin = pd.read_excel('../01_data_get/Data-人民日报.xlsx',header=0)
df_yangshi = pd.read_excel('../01_data_get/Data-央视新闻.xlsx',header=0)
```

```python
df_renmin
```

```python
# 获取两个媒体的播放量数据
watch_renmin = df_renmin['播放量(万)']
watch_yangshi = df_yangshi['播放量(万)']
```

```python
# 2. 计算平均播放量


# 计算总和
watchnum_renmin = 0
watchnum_yangshi = 0
for watch in watch_renmin:
    watchnum_renmin += watch
for watch in watch_yangshi:
    watchnum_yangshi += watch

# 计算平均值
a_renmin = watchnum_renmin*10000/len(watch_renmin)
a_yangshi = watchnum_yangshi*10000/len(watch_yangshi)

```

```python
# 3. 可视化
# 处理中文字体的显示问题
my_font = font_manager.FontProperties(fname="C:/WINDOWS/Fonts/STSONG.TTF")
#定义中文字体属性，文字储存路径可以在C:/WINDOWS/Fonts/找到，这里设置为宋体

x = ['人民日报','央视新闻']
y = [a_renmin,a_yangshi]


plt.bar(x,y,color='royalblue',alpha=0.7,width=0.5) # 设置变量、颜色、遮罩（不透明度）

plt.legend(loc="upper right",prop=my_font)  # 设置位置以及中文字体

# 坐标轴名称的设置
plt.ylabel('播放量',fontproperties = my_font,fontsize = 14)

# 图片的标题设置
plt.title(u'人民日报与央视新闻平均播放量直方图',fontproperties = my_font,fontsize = 14)

# x轴刻度设置
plt.xticks(x, fontproperties=my_font,fontsize = 10)

# plt.show() # 图片显示，和savefig两个写一个就好
plt.savefig(r'人民日报与央视新闻平均播放量.jpg')  # 图片保存
```
